```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Time Series

## Visualization of Trend, Seasonality and Noise

```python
# We simulate the trend
trend = 2.2 * np.array(range(1095)) + 5

# We create our series describing the periodic component 
period = np.array([
    160 * np.sin(2 * np.pi / 365 * x) + 80 * np.cos(2 * np.pi / 100 * x)
    for x in range(0, 1095)
])

# We simulate noise 
white_noise = 10*np.random.normal(3, 2, size=1095)

print('Trend, seasonality and noise successfully created.')
```

```python
fig, ax = plt.subplots(1, 3, figsize=(16, 4))

# Trend
plt.subplot(1, 3, 1)
plt.title("Trend")
plt.plot(trend, color = "#45B39D")

# Seasonality
plt.subplot(1, 3, 2)
plt.title("Seasonality")
plt.plot(period, color = "#C70039")

# White noise
plt.subplot(1, 3, 3)
plt.title("Noise")
plt.plot(white_noise, color = "#34495E")

# Affichage
plt.show();
```

### Additive Model

```python
fig, ax = plt.subplots(1, 1, figsize=(16, 4))

plt.subplot(1, 1, 1)
plt.title("Combine")
plt.plot(trend+period+white_noise, color = "#34495E")

plt.plot(trend+period, color = "#C70039")

plt.plot(trend, color = "#45B39D")

plt.show();
```

### Multiplicative Model

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 4))

plt.subplot(1, 2, 1)
plt.plot(trend + period+ white_noise, color="#808B96")
plt.title("Time Series : Additive model")

plt.subplot(1, 2, 2)
plt.plot(trend * (1 + period) * (1 + white_noise), color="#808B96")
plt.title("Time series : Multiplicative model");
```
